# **Prodotto di due radicali con stesso indice**

Scriviamo un programma che:



###1) Chieda in input l'indice comune di due radicali

```python
indice_comune = int(input("Inserisci l'indice comune"))
```

###2) Controlli che l'indice sia un numero naturale diverso da 0; in caso contrario, si fermi, dopo aver dato un messaggio di errore, e invita a cambiarlo;

```python
while (indice_comune <=0):
  #if (indice_comune <= 0):
  print("ERRORE: l'indice comune deve essere un numero naturale diverso da 0")
  indice_comune = int(input("Inserisci l'indice comune"))
```

###3) Chieda in input i due radicandi, limitandoci a numeri interi;

```python
radicando1 = int(input("Inserisci 1o radicando"))
if (indice_comune % 2)==0:  # Verifico se l'indice è pari
  while(radicando1 < 0):    # Controllo C.E. radicando
    print("ERRORE: il radicando non può essere negativo")
    radicando1 = int(input("Inserisci 1o radicando"))

radicando2 = int(input("Inserisci 2o radicando"))
if (indice_comune % 2)==0:
  while(radicando2 < 0):    # Controllo C.E. radicando
    print("ERRORE: il radicando non può essere negativo")
    radicando2 = int(input("Inserisci 2o radicando"))
```

###4) Calcoli il prodotto dei radicandi

```python
prodotto_radicandi = radicando1 * radicando2
```

###5) Fornisca in output l'indice e il radicando del prodotto.

```python
print(f'Il risultato è: {prodotto_radicandi}**(1/{indice_comune})')
```
